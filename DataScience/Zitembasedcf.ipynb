```python
import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('C:/Users/zliu/Documents/GitHub/udemy_dsmlwp/datascience/ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3))

m_cols = ['movie_id', 'title']
movies = pd.read_csv('C:/Users/zliu/Documents/GitHub/udemy_dsmlwp/datascience/ml-100k/u.item', sep='|', names=m_cols, usecols=range(2))

ratings = pd.merge(movies, ratings)

ratings.head()
```

```python
userRatings = ratings.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
userRatings.head()
```

```python
corrMatrix = userRatings.corr(min_periods=100)
corrMatrix.head()
```

```python
myRating = userRatings.iloc[0].dropna()
myRating
```

```python
simiList = pd.Series()
for i in range(len(myRating.index)):
    movTitle = myRating.index[i]
    movSimi = corrMatrix[movTitle].dropna()
    movSimi = movSimi.map(lambda x: x * (myRating[i] - myRating.max()/2))
    simiList = simiList.append(movSimi)
simiList = simiList.groupby(simiList.index).sum()
simiList.sort_values(inplace = True, ascending = False)
```

```python
recList = simiList.drop(myRating.index)
recList = recList[:20]
recList
```

```python
userStat = pd.DataFrame(columns = ('total','aveRating'))
for i in range(len(userRatings.index)):
    stat = [userRatings.iloc[i].notnull().sum(), userRatings.iloc[i][userRatings.iloc[i].notnull()].mean()]
    userStat.loc[i] = stat
userStat.head()
```
